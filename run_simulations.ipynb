```python
from district import District
from city import City
from schelling import Schelling
from gaylord import Gaylord
from visualization import Visualization
```

```python
district = District()
print(f'Population features: \nrases: {district.rases} \nmaterial status: {district.salaries}')
```

### Segregation using Schelling model

#### Testing agents_ratio parameter (0.6, 0.7, 0.5) on simulation of the algorithm

```python
city = City(district, size=80, agents_ratio=0.6)
schelling = Schelling(city, district, a=4, n_steps=8000)

visualization = Visualization(schelling)
visualization.show_results()
```

```python
city = City(district, size=80, agents_ratio=0.7)
schelling = Schelling(city, district, a=4, n_steps=8000)

visualization = Visualization(schelling)
visualization.show_results()
```

```python
city = City(district, size=80, agents_ratio=0.5)
schelling = Schelling(city, district, a=4, n_steps=8000)

visualization = Visualization(schelling)
visualization.show_results()
```
